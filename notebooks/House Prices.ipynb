```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import pickle
```

# Dados

```python
uri = '/content/drive/MyDrive/Estudos/Data_Science/Alura/Python/MLOps/dados/casas.csv'
dados = pd.read_csv(uri)
```

```python
dados.head()
```

# Tratamentos

```python
dados.head()
```

```python
X = dados.drop('preco', axis='columns')
y = dados['preco']
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Modelo

```python
modelo = LinearRegression()
modelo.fit(x_treino, y_treino)
```

```python
(modelo.coef_, modelo.intercept_)
```

```python
modelo.predict([[120, 2001, 2]])
```

#Salvando Modelo

```python
caminho_salvar = '/content/drive/MyDrive/Estudos/Data_Science/Alura/Python/MLOps/modelos/'
nome_arquivo = 'modelo.sav'
pickle.dump(modelo, open(caminho_salvar+nome_arquivo, 'wb'))
```
